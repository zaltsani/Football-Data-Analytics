```python
Home_Team = 'Roma'
Away_Team = 'Milan'
```

## Home Data

```python
Summary = match[3]
Passing = match[4]
Pass_Types = match[5]
Defensive_Actions = match[6]
Possession = match[7]
Miscellaneous_Stats = match[8]
```

### Passing

```python
Passing = Passing.rename(columns={'Cmp': 'Total Passes Completed'})
Passing = Passing.rename(columns={'Att': 'Total Passes Attempted'})
Passing = Passing.rename(columns={'Cmp%': 'Total Passes Completion%'})
Passing = Passing.rename(columns={'TotDist': 'Total Passing Distance'})
Passing = Passing.rename(columns={'PrgDist': 'Total Progressive Passing Distance'})
Passing = Passing.rename(columns={'Cmp.1': 'Short Passes Completed'})
Passing = Passing.rename(columns={'Att.1': 'Short Passess Attempted'})
Passing = Passing.rename(columns={'Cmp%.1': 'Short Passes Completion%'})
Passing = Passing.rename(columns={'Cmp.2': 'Medium Passes Completed'})
Passing = Passing.rename(columns={'Att.2': 'Medium Passes Attempted'})
Passing = Passing.rename(columns={'Cmp%.2': 'Medium Passes Completion%'})
Passing = Passing.rename(columns={'Cmp.3': 'Long Passes Completed'})
Passing = Passing.rename(columns={'Att.3': 'Long Passes Attempted'})
Passing = Passing.rename(columns={'Cmp%.3': 'Long Passes Completion%'})
```

### Pass Types

```python
# Total Passes Attempted Has the Same Data in Passing Data
Pass_Types = Pass_Types.drop(columns='Att')

# Pass Types
Pass_Types = Pass_Types.rename(columns={'Live': 'Live-ball Passes'})
Pass_Types = Pass_Types.rename(columns={'Dead': 'Dead-ball Passes'})
Pass_Types = Pass_Types.rename(columns={'FK': 'Passes from Free Kicks'})
Pass_Types = Pass_Types.rename(columns={'TB': 'Through Balls'})
Pass_Types = Pass_Types.rename(columns={'Sw': 'Switches'})
Pass_Types = Pass_Types.rename(columns={'Crs': 'Crosses'})
Pass_Types = Pass_Types.rename(columns={'TI': 'Throw-ins Taken'})
Pass_Types = Pass_Types.rename(columns={'CK': 'Corner Kicks'})

# Corner Kicks
Pass_Types = Pass_Types.rename(columns={'In': 'Inswinging Corner Kicks'})
Pass_Types = Pass_Types.rename(columns={'Out': 'Outswinging Corner Kicks'})
Pass_Types = Pass_Types.rename(columns={'Str': 'Straight Corner Kicks'})

# Outcomes
Pass_Types = Pass_Types.rename(columns={'Cmp': 'Outcome Passes Completed'})
Pass_Types = Pass_Types.rename(columns={'Off': 'Passes Offside'})
Pass_Types = Pass_Types.rename(columns={'Blocks': 'Passes Blocked'})
```

### Defensive Actions

```python
Defensive_Actions = Defensive_Actions.rename(columns={'Tkl': 'Tackles'})
Defensive_Actions = Defensive_Actions.rename(columns={'TklW': 'Tackles Won'})
Defensive_Actions = Defensive_Actions.rename(columns={'Def 3rd': 'Tackles (Def 3rd)'})
Defensive_Actions = Defensive_Actions.rename(columns={'Mid 3rd': 'Tackles (Mid 3rd)'})
Defensive_Actions = Defensive_Actions.rename(columns={'Att 3rd': 'Tackles (Att 3rd)'})
Defensive_Actions = Defensive_Actions.rename(columns={'Tkl.1': 'Dribblers Tackled'})
Defensive_Actions = Defensive_Actions.rename(columns={'Att': 'Dribbles Challenged'})
Defensive_Actions = Defensive_Actions.rename(columns={'Tkl%': '% of Dribblers Tackled'})
Defensive_Actions = Defensive_Actions.rename(columns={'Lost': 'Challenges Lost'})
Defensive_Actions = Defensive_Actions.rename(columns={'Sh': 'Shots Blocked'})
Defensive_Actions = Defensive_Actions.rename(columns={'Pass': 'Passes Blocked'})
Defensive_Actions = Defensive_Actions.rename(columns={'Int': 'Interception'})
Defensive_Actions = Defensive_Actions.rename(columns={'Clr': 'Clearances'})
Defensive_Actions = Defensive_Actions.rename(columns={'Err': 'Errors'})
```

### Possession

```python
# Touches
Possession = Possession.rename(columns={'Def Pen': 'Touches (Def Pen)'})
Possession = Possession.rename(columns={'Def 3rd': 'Touches (Def 3rd)'})
Possession = Possession.rename(columns={'Mid 3rd': 'Touches (Mid 3rd)'})
Possession = Possession.rename(columns={'Att 3rd': 'Touches (Att 3rd)'})
Possession = Possession.rename(columns={'Att Pen': 'Touches (Att Pen)'})
Possession = Possession.rename(columns={'Live': 'Touches (Live-Ball)'})

# Take-Ons
Possession = Possession.rename(columns={'Att': 'Take-Ons Attempted'})
Possession = Possession.rename(columns={'Succ': 'Succesful Take-Ons'})
Possession = Possession.rename(columns={'Succ%': 'Succesful Take-On %'})
Possession = Possession.rename(columns={'Tkld': 'Times Tackled During Take-On'})
Possession = Possession.rename(columns={'Tkld%': 'Tackled During Take-On Percentage'})

# Carries
Possession = Possession.rename(columns={'TotDist': 'Total Carrying Distance'})
Possession = Possession.rename(columns={'PrgDist': 'Progressive Carrying Distance'})
Possession = Possession.rename(columns={'PrgC': 'Progressive Carries'})
Possession = Possession.rename(columns={'1/3': 'Carries into Final Third'})
Possession = Possession.rename(columns={'CPA': 'Carries into Penalty Area'})
Possession = Possession.rename(columns={'Mis': 'Miscontrols'})
Possession = Possession.rename(columns={'Dis': 'Dispossessed'})

# Receiving
Possession = Possession.rename(columns={'Rec': 'Passes Received'})
Possession = Possession.rename(columns={'PrgR': 'Progressive Passes Received'})
```

### Miscellaneous Stats

```python
# Performance
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'CrdY': 'Yellow Cards'})
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'CrdR': 'Red Cards'})
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'2CrdY': 'Second Yellow Card'})
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'Fls': 'Fouls Commited'})
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'Fld': 'Fouls Drawn'})
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'Off': 'Offsides'})
Miscellaneous_Stats = Miscellaneous_Stats.drop(columns='Crs')
Miscellaneous_Stats = Miscellaneous_Stats.drop(columns='Int')
Miscellaneous_Stats = Miscellaneous_Stats.drop(columns='TklW')
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'PKwon': 'Penalty Kicks Won'})
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'PKcon': 'Penalty Kicks Conceded'})
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'OG': 'Own Goals'})
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'Recov': 'Ball Recoveries'})

# Aerial Duels
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'Won': 'Aerials Won'})
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'Lost': 'Aerials Lost'})
Miscellaneous_Stats = Miscellaneous_Stats.rename(columns={'Won%': '% of Aerials Won'})
```

### Merge

```python
Home_Player_Stats = pd.merge(Passing, Pass_Types, on=['Player', '#', 'Nation', 'Pos', 'Age', 'Min'])
Home_Player_Stats = pd.merge(Home_Player_Stats, Defensive_Actions, on=['Player', '#', 'Nation', 'Pos', 'Age', 'Min'])
Home_Player_Stats = pd.merge(Home_Player_Stats, Possession, on=['Player', '#', 'Nation', 'Pos', 'Age', 'Min'])
Home_Player_Stats = pd.merge(Home_Player_Stats, Miscellaneous_Stats, on=['Player', '#', 'Nation', 'Pos', 'Age', 'Min'])
```

```python
Home_Player_Stats = Home_Player_Stats.drop(len(Home_Player_Stats) - 1)
```

```python
Home_Player_Stats = Home_Player_Stats.rename(columns={'#': 'Player Number'})
```

```python
print(Home_Player_Stats.to_markdown())
```
