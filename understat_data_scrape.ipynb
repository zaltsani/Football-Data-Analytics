# Scrape Match

## Import Packages

```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

## Input Match ID

```python
match_id = input("Match ID: ")
```

## Scraping Procces

```python
#scrape a single game shots. We will allow users to enter a base url
base_url = 'https://understat.com/match/'
url = base_url + str(match_id)
```

```python
#Use requests to get the webpage and BeautifulSoup to parse the page
res = requests.get(url)
soup = BeautifulSoup(res.content, 'lxml')
scripts = soup.find_all('script')
```

```python
html = BeautifulSoup(res.content, 'html')
```

```python
# Get Title
title = str(soup.find('title'))
split_start = '<title>'
split_end = ' |'
title = title.split(split_start, 1)[1].split(split_end, -1)[0]
title = title.replace(' ', '_').lower()
```

```python
#get only the shotsData
strings = scripts[1].string
```

```python
# strip unnecessary symbols and get only JSON data 
ind_start = strings.index("('")+2 
ind_end = strings.index("')") 
json_data = strings[ind_start:ind_end] 
json_data = json_data.encode('utf8').decode('unicode_escape')

#convert string to json format
data = json.loads(json_data)
```

## Data Processing

```python
head = []
dict = {}

for _ in data['h'][0].keys():
    head.append(_)

for k in head:
        dict[k] = []

for ha in data.keys():
    for i in range(len(data[ha])):
        for k in head:
            dict[k].append(data[ha][i][k])

df_data = pd.DataFrame(dict)
df_data
```

## Save Data into CSV

```python
import os
path = r'D:\Football\Data\shot_stats'
```

```python
df_data.to_csv(os.path.join(path, f'shot_stat_{title}.csv'))
```

# Scrape Team and Player

## Import Packages

```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import os
```

## Input

```python
team = 'Wolverhampton_Wanderers'
year = 2023
```

## Scraping Procces

```python
base_url = 'https://understat.com/team/'
url = base_url + str(team) + '/' + str(year)
url
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.content, 'lxml')
scripts = soup.find_all('script')
```

```python
html = BeautifulSoup(res.content, 'html')
```

```python
# Get Title
title = str(soup.find('title'))
split_start = '<title>'
split_end = ' |'
title = title.split(split_start, 1)[1].split(split_end, -1)[0]
title = title.replace(' ', '_').lower()
title
```

```python
# Statistics Data
statistics_datas = scripts[2].string
```

```python
# Player Data
players_datas = scripts[3].string
```

```python
# strip unnecessary symbols and get only JSON data 
ind_start = statistics_datas.index("('")+2 
ind_end = statistics_datas.index("')") 
json_data = statistics_datas[ind_start:ind_end] 
json_data = json_data.encode('utf8').decode('unicode_escape')

#convert string to json format
statistic_data = json.loads(json_data)
```

```python
# strip unnecessary symbols and get only JSON data 
ind_start = players_datas.index("('")+2 
ind_end = players_datas.index("')") 
json_data = players_datas[ind_start:ind_end] 
json_data = json_data.encode('utf8').decode('unicode_escape')

#convert string to json format
player_data = json.loads(json_data)
```

## Data Processing: Team

```python
section = []
for i in statistic_data.keys():
    section.append(i)
```

```python
path = 'D:\\Football\\My_Code\\webscraping_template\\' + team + "_" + str(year)
```

```python
for i in section:
    sect = statistic_data[i]
    dict = {}
    head = []
    sub_head = []

    # Create Head
    for j in sect.keys():
        head.append(j)

    # Create Sub_Head
    for k in sect[head[0]].keys():
        if str(type(sect[head[0]][k])) == "<class 'dict'>":
            for l in sect[head[0]][k].keys():
                sub_head.append(f"{l}_against")
        else:
            sub_head.append(k)

    # If Situation, Input Situation Value
    if i == 'situation':
        dict['situation'] = head

    # Create Dictionary Keys
    for m in sub_head:
        dict[m] = []

    # Input Value Sub_Head
    for n in head:
        head_data = sect[n]
        for o in sub_head:
            if o not in head_data.keys():
                against_data = head_data['against']
                p = o.split('_', 1)[0]
                dict[o].append(against_data[p])
            else:
                dict[o].append(head_data[o])
    
    df = pd.DataFrame(dict)

    df.to_csv(os.path.join(path, f'{i}.csv'))
```
